### Ánalisis exploratorio de mi cuenta personal en Netflix

```python
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('NetflixViewingHistory.csv')
df
```

```python
df.info()

print('\n')
df.shape
```

Se ve que no hay valores que falten, lo que facilitará el análisis

```python
#1
# Haré una lista de los títulos ya que será mejor para el tratamiento de datos
# separarlo del símbolo ':'

list_sep = df['Title'].str.split(pat = ':', expand=False).to_frame()
list_sep
```

```python
#2
# vemos el número de elementos de cada título


list_sep['num_part'] = list_sep['Title'].apply(len)
list_sep
```

```python
# contamos el número de elementos que tiene cada título

list_sep['num_part'].value_counts()
```

```python
#Creamos una nueva columna donde colocaremos el tipo
# en función del número de partes, ya que los episodios de series
# suelen tener un título más largo

list_sep['Tipo'] = np.where(list_sep['num_part'] < 3, 'Pelicula', 'Serie')
list_sep
```

```python
#lo unimos al dataframe original

df = pd.concat([df,list_sep['Tipo']], axis=1)
df
```

```python
#Añadimos nuevas columnas en función de las partes de los títulos
# así el análisis exploratorio se enriquece

cols_sep = df['Title'].str.split(pat=':', expand=True)
cols_sep.columns =['part1', 'part2', 'part3', 'part4', 'part5']
cols_sep
```

```python
#lo unimos al dataframe principal

df = pd.concat([df,cols_sep],axis=1)
df
```

```python
#dado que la columna índice es irrelevante para el análisis
#la reemplazamos por la fecha, convirtiéndolo en formato fecha

df['Fecha'] = pd.to_datetime(df['Date'])
df.set_index('Fecha',inplace=True)
df
```

```python
#con una función que tome el año, mes y día de la semana de cada título
# creamos nuevas columnas
'''
se crean nuevas columnas, en la que cada una de ellas
toma el valor correspondiente según su índice,
es decir, el primer índice es el año, tal como puede verse
en la columna 'Fecha', y así sucesivamente. 

Por tanto, tomando el índice de cada fecha, lo almacenamos
en columnas nuevas


'''

def fechas(data):
    data['año'] = data.index.year
    data['mes'] = data.index.month_name()
    data['dia_mes'] = data.index.day
    data['dia_semana'] = data.index.day_name()
    
    return(data)

fechas(df)
```

```python
#por comodidad, reemplazamos los nombres en inglés por español
# con el método replace de pandas


to_rep1 = {'Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'}

to_rep2 = {'January','February','March','April','May','June',
          'July','August','September','October','November','December'}

to_rep3 = {'Title'}

vals1 = {'Lunes','Martes','Miercoles','Jueves','Viernes','Sabado','Domingo'}

vals2 = {'Enero','Febrero','Marzo','Abril','Mayo','Junio',
        'Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre'}

vals3 = {'Titulo'}


df.replace(to_replace=to_rep1, value=vals1, inplace=True)
df.replace(to_replace=to_rep2, value=vals2, inplace=True)
df.replace(to_replace=to_rep3, value=vals3, inplace=True)

df
```

```python
#importamos la librería datetime
# con el objetivo de calcular cuántos días
# he consumido usando Netflix


from datetime import date

actual = pd.Timestamp(date.today())

dia_uno = df.index.min()

total = actual - dia_uno

print(f'He estado usando Netflix durante {total.days} dias')
```

```python
#con una función pasamos los días a años

'''
tomamos el resultado anterior en días y lo dividimos en 365
para saber los años. El resto serán en días, que lo almacenamos
en una nueva variable y la redondeamos

'''

def tiempo(total):
    years = total // 365
    rem_year = total%365 #resto es 22 dias
    dias = round(rem_year,2)
    
    
    total = print(f'He estado usando Netflix durante {years} años y {dias} dias' )
    
    return total

tiempo(total.days)
```

```python
#Contemos el número de días que he estado viendo series y películas
#de mi vida, sin parar. Haciendo una media de tres por semana

min_serie = 50
min_peli = 120

consumo = df.loc[df.año < 2023].groupby('Tipo').Title.count()

minutos_pelis_año = consumo['Pelicula'] * min_peli / 3

minutos_series_año = consumo['Serie'] * min_serie / 3

dias_pelis_año = minutos_pelis_año / 60 / 24

dias_series_año = minutos_series_año / 60 / 24

print(f'Al año dedico {round(dias_series_año)} dias de mi vida a ver series y {round(dias_pelis_año)} dias de mi vida a ver pelis')


```
